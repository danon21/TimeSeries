```python
import os
from os import path
import numpy as np
import matplotlib as mp
from matplotlib import pyplot as plt
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dirname = os.path.abspath(os.curdir) + r'\data\underwork\5'
tsdf_c = pd.read_csv(path.join(os.sep, dirname, 'calm_p.csv'))
tsdf_c.set_index('Time').sort_index()

psg = pd.read_csv(os.path.abspath(os.curdir)+ r'\data\passengers.csv')
```

```python
fig, aws = plt.subplots(2, 1, figsize=(20,15))
aws[0].plot(psg['Passengers'])
aws[1].plot(tsdf_c['1'])
```

```python
noise = np.random.normal(0,1,1000)
decompose = seasonal_decompose(noise, period=50, model="additive")
decompose.plot()
plt.show()
```
